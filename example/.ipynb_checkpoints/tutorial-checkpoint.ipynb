```python
# load basic packages and import the supervoxel package
import numpy as np
import sys
sys.path.append('../src')
import matplotlib.pyplot as plt
from supervoxel import *
from vis_helper import *
plt.rcParams.update({'font.size': 16})
```

```python
# fine scale image
n   = np.array([160, 160, 160])       # nz, ny, nx
L   = np.array([1.0, 1.0, 1.0])    # Lz, Ly, Lx
img = np.zeros(n,dtype=np.uint8)
# coordinates along x-y-z-axis
x   = L[2]*(np.linspace(0,1.0,n[2]+1)+0.5/n[2]-0.5)[:-1]
y   = L[1]*(np.linspace(0,1.0,n[1]+1)+0.5/n[1]-0.5)[:-1]
z   = L[0]*(np.linspace(0,1.0,n[0]+1)+0.5/n[0]-0.5)[:-1]
```

## Example: Centered Sphere

- place a sphere with radius $r > 0$ in the fine scale image
- show slices of the image
- coarse-grain $\blacktriangleright$ super-boxels
- show slices of the coarse-grained image
- visualize select 2D projefctions of the normals

```python
n_vx= [5] * 3
print('fine scale image resolution         ', n)
print('RVE dimensions                      ', L)
print('supver-boxel size                   ', n_vx)
n_coarse = []
for (i,j) in zip(n,n_vx):
    n_coarse = n_coarse + [int(i/j)]
print('coarse grained image resolution     ', n_coarse)
```

```python
r   = 0.35         # the radius
V_s = 4./3.*np.pi*r*r*r
A_s = np.pi*r*r*4.
V   = L.prod()
c   = V_s/V      # volume fraction
print('sphere of radius ', r, 'volume fraction %5.2f %%' % (100*c))

# generate actual image
img = (((z**2)[:,None,None]+(y**2)[None,:,None]+(x**2)[None,None,:])<r*r)
```

```python
show_microstructure(img, L=L, idx=[0, int(n[0]/2)])
```

### Perform the super-voxel/boxel-ation

```python
verbose = True
vol_frac, normals, normals_classic, combi_img =  supervoxelize_image( img, L=L, voxel_size=n_vx, \
    hdfname='sphere.h5', overwrite=True, gzip=True, quiet=not verbose )
```

### Showing some slices of the coarse grained composite discretization
* visualize the distribution (inclusion vs. composite voxels)
* including normal vectors *(planar projection only; see [paraview tutorial](#paraview) on how to perform 3D visualization)*


```python
show_slices( vol_frac=vol_frac, combi_img =combi_img, normals=normals, L=L, idx=[ int(n[0]/4), int(n[0]/2) ])
```

## Example: Prism

- place a triangular prism of length $l > 0$ in the fine scale image
- show slices of the image
- coarse-grain $\blacktriangleright$ super-boxels
- show slices of the coarse-grained image
- compare the normals using the new approach from [\[2\]](#ref:keshav2022) to the established one from [\[1\]](#ref:kabel2015) 

```python
l   = 0.5
# draw a prism:
img = (np.abs(z[:,None,None]) <= 0.5*l ) \
    * ((y[None,:,None]+0.75*x[None,None,:])< 0.2) \
    * ((0.5*y[None,:,None]+1.5*x[None,None,:])> -0.2) \
    * ((y[None,:,None]-1.75*x[None,None,:]) > -0.2) 
```

```python
show_microstructure(img, L = L, idx=[1,int(n[0]/2)])
```

```python
vol_frac, normals, normals_classic, combi_img =  supervoxelize_image( img, L=L, voxel_size=n_vx, \
    hdfname='prism.h5', groupname='test', overwrite=True, gzip=True, quiet=True )
```

```python
show_slices( vol_frac=vol_frac, combi_img=combi_img, normals=normals, L=L, idx=[1,int(n[0]/2)] )
show_slices( vol_frac=vol_frac, combi_img=combi_img, normals=normals_classic, L=L, idx=[1,int(n[0]/2)] )
```

### Step-by-step tutorial
1. launch `paraview`
2. select `File->Open` or click
<img src="gfx/step_file_open.png" width="35%"></img>
3. select the two `xdmf` files (multiple files can be selected using `shift`- or `ctrl/CMD`-click; select `XDMF Reader` from the context menu and click `Ok`
4. in the properties context menu (usually bottom left) click `apply`
<img src="gfx/step_apply_file_open.png" width="50%"></img>

### Thresholding
1. select one of the files in the `pipeline browser` (usually placced in the top left sidebar)
2. select `Filters->Common->Threshold` from the menu or click
<img src="gfx/step_threshold.png" width="50%"></img>
3. select `Materials` (for the high-resolution image) or `COMBIMAT` (for the composite dataset)
<img src="gfx/step_threshold_param1.png" width="50%"></img>
4. adjust range (0: matrix, 1: inclusion, 2: composite voxels) in order to visualize different parts of the domain; e.g., stripping the matrix phase is achieved by setting the lower threshold value to 0.5

### Expected outcome
<img src="./gfx/step_sphere_threshold.png" width="40%"></img>

### Visualization of the normal vectors

1. after thresholding, select the thresholded domain containing only composite voxels (threshold `COMBIMAT` using, e.g., minimum value 1.5, maximum value 2 for the composite dataset) in the `pipeline browser`
<img src="gfx/step_threshold_param2.png" width="50%"></img>
2. select `Filters->Common->Glyph` or click
<img src="gfx/step_glyph.png" width="25%"></img>
3. select `Orientation Array` either `NORMAL` or `NORMAL_CLASSIC`
<img src="gfx/step_glyph_param1.png" width="50%"></img>

<img src="gfx/step_glyph_param2.png" width="50%"></img>
5. select `Glyph Mode->All Points`
<img src="gfx/step_glyph_param3.png" width="50%"></img>
6. click `Apply`
7. scroll further down in the properties dialog and select `Coloring->Solid Color` (actual color can be modified via `Edit`)
<img src="gfx/step_glyph_param4.png" width="50%"></img>
